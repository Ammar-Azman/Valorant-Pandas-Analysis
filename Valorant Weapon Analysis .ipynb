# Average Damage 

```python
Type_weapon = input("Insert your weapon type to be analyze:")

if Type_weapon == "smg":
    Type_weapon = Type_weapon.upper()
else:
    Type_weapon = Type_weapon.capitalize()

#data set edited by taking average value
df["Avg HDMG"] = ((df.HDMG_0 + df.HDMG_1 + df.HDMG_2)/3).round(decimals=2)
df["Avg BDMG"] = ((df.BDMG_0 + df.BDMG_1 + df.BDMG_2)/3).round(decimals=2)
df["Avg LDMG"] = ((df.LDMG_0 + df.LDMG_1 + df.LDMG_2)/3).round(decimals=2)

#for graph

df_1 = df[["Weapon Type","Name", "Avg HDMG","Avg BDMG", "Avg LDMG"]]
df_1 = df.loc[df["Weapon Type"] == Type_weapon]
df_1 = df_1.sort_values(["Avg HDMG","Avg BDMG", "Avg LDMG"], ascending=False)

#for table

df_t = df[["Weapon Type","Name", "Avg HDMG","Avg BDMG", "Avg LDMG"]]
df_t = df_t.loc[df["Weapon Type"] == Type_weapon].sort_values(["Avg HDMG"], ascending= False)
df_t = df_t.groupby(["Weapon Type", "Name"])
df_t = df_t.first()

# bar plot
ax = plt.figure().add_axes([0,0,2.5,1])
ax.bar(df_1.Name, df_1["Avg HDMG"])
ax.bar(df_1.Name, df_1["Avg BDMG"])
ax.bar(df_1.Name, df_1["Avg LDMG"])
plt.title(Type_weapon + " Average Damage")
plt.xlabel("Weapon Name")
plt.ylabel("Average Damage")
plt.legend(["Head", "Body", "Leg"])
plt.show()
df_t
```

# Weapon Headshot Analysis 

```python
Type_weapon = input("Insert your weapon type to be analyze:")

if Type_weapon == "smg":
    Type_weapon = Type_weapon.upper()
else:
    Type_weapon = Type_weapon.capitalize()

val_hs = df[["Name", "Weapon Type",'HDMG_0', 'HDMG_1', 'HDMG_2' ]]
val_hs = val_hs.loc[val_hs["Weapon Type"] == Type_weapon]
val_hs = val_hs.sort_values(['HDMG_0', 'HDMG_1', 'HDMG_2'], ascending = False)


import matplotlib.pyplot as plt

ax = plt.figure().add_axes([0,0,2.5,1])

ax.bar(val_hs.Name, val_hs.HDMG_0)
ax.bar(val_hs.Name, val_hs.HDMG_1)
ax.bar(val_hs.Name, val_hs.HDMG_2)
plt.title(Type_weapon + " Headshot Damage by Distance")
plt.legend(["Close", "Medium", "Far"])
plt.xlabel("Weapon")
plt.ylabel("Headshot Damage")
plt.show()
val_hs.reset_index(drop=True)
```

# Fire rate Analysis

```python
Type_weapon = input("Insert your weapon type to be analyze:")

if Type_weapon == "smg":
    Type_weapon = Type_weapon.upper()
else:
    Type_weapon = Type_weapon.capitalize()

# filtering Fire Rate only
val_rat = df[["Name", "Weapon Type", "Fire Rate" ]]

# choosing weapon Type
val_rat = val_rat.loc[val_rat["Weapon Type"] == Type_weapon]
val_rat = val_rat.sort_values(["Fire Rate"], ascending = False)
print(val_rat)

import matplotlib.pyplot as plt

ax = plt.figure().add_axes([0,0,2.5,1])

ax.bar(val_rat.Name, val_rat["Fire Rate"])
plt.xlabel("Weapon")
plt.ylabel("Fire Rate")
plt.show()
```
